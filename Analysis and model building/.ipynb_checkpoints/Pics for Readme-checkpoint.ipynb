```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import pickle
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
churn_df_details = pickle.load(open('churn_df_details.pkl','rb'))

df = churn_df_details['df']
obj_cols = churn_df_details['obj_cols']
num_cols = churn_df_details['num_cols']
```

```python
plt.rcParams['font.size'] = '14'
```

### 1) Tenure vs Churn

```python
plt.figure(figsize=(12,4))
sns.boxplot(x=df['tenure'],y=df['Churn'])
plt.title('Tenure vs Churn')
plt.savefig('Tenure_vs_Churn.jpg')
plt.show()
```

### 2) Contract vs Churn

```python
plt.figure(figsize=(12,8))
ax1, ax2 = plt.subplot(2,1,1), plt.subplot(2,1,2)

cross_tab_prop = pd.crosstab(df['Contract'],df['Churn'])
cross_tab_prop.plot(kind='barh',stacked=True,ax=ax1)
ax1.legend(loc="center",ncol=1,fontsize='x-small',bbox_to_anchor=(1.05, 0.9))
ax1.set_title(f'Contract type vs Churn')
ax1.set_xlabel(f'Churn')

cross_tab_prop = pd.crosstab(df['Contract'],df['Churn'],normalize='index')
cross_tab_prop.plot(kind='barh',stacked=True,ax=ax2)
ax2.legend(loc="center",ncol=1,fontsize='x-small',bbox_to_anchor=(1.05, 0.9))
ax2.set_title(f'Contract type vs Churn ratio')
ax2.set_xlabel(f'Churn ratio')

plt.tight_layout()
plt.savefig('ContractType_vs_Churn.jpg')
plt.show()
```

### 5) OnlineSecurity vs Churn

```python
plt.figure(figsize=(18,12))
ax1, ax2, ax3, ax4, ax5, ax6 = plt.subplot(3,2,1), plt.subplot(3,2,2), plt.subplot(3,2,3), plt.subplot(3,2,4),\
                                plt.subplot(3,2,5), plt.subplot(3,2,6)
axes = [ax1, ax2, ax3, ax4, ax5, ax6]
cols = ['InternetService','TechSupport','OnlineSecurity','OnlineBackup','StreamingTV','StreamingMovies']

for ax, col in zip(axes,cols):
    cross_tab_prop = pd.crosstab(df[col],df['Churn'])
    cross_tab_prop.plot(kind='barh',stacked=True,ax=ax)
    ax.legend(loc="center",ncol=1,fontsize='x-small',bbox_to_anchor=(1.1, 0.9))
    ax.set_title('{} vs Churn'.format(col))
    ax.set_xlabel(f'Churn')

plt.tight_layout()
plt.savefig('Internet-PremiumServices_vs_Churn.jpg')
plt.show()
```

## Final model performance

```python
churn_df_details = pickle.load(open('churn_df_details.pkl','rb'))
X, Y = churn_df_details['df_final'], churn_df_details['y']['Churn'].values

model = pickle.load(open('model.pkl', 'rb'))
ypred = model.predict(X)

model_perf = pd.DataFrame({'Actual':Y,'Predicted':ypred})
```

```python
plt.figure(figsize=(12,8))
ax1, ax2, ax3, ax4 = plt.subplot(2,2,1), plt.subplot(2,2,2), plt.subplot(2,2,3), plt.subplot(2,2,4)
sns.heatmap(pd.crosstab(model_perf['Predicted'],model_perf['Actual']),ax=ax1,annot=True,fmt='.0f',cmap='Blues')
sns.heatmap(pd.crosstab(model_perf['Predicted'],model_perf['Actual'],normalize='all'),ax=ax2,annot=True,fmt='.2f',cmap='Blues')
sns.heatmap(pd.crosstab(model_perf['Predicted'],model_perf['Actual'],normalize='index'),ax=ax3,annot=True,fmt='.2f',cmap='Blues')
sns.heatmap(pd.crosstab(model_perf['Predicted'],model_perf['Actual'],normalize='columns'),ax=ax4,annot=True,fmt='.2f',cmap='Blues')
ax1.set_title('Crosstab in numbers')
ax2.set_title('Crosstab in % of total')
ax3.set_title('Precision')
ax4.set_title('Recall')
plt.tight_layout()
plt.savefig('Final_model_performance_crosstab.jpg')
plt.show()
```
